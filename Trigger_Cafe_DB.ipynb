```python
import psycopg2
import os

# Verbindung zur PostgreSQL-Datenbank herstellen
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    user="postgres",
    password=os.environ['DB_PASSWORD'],
    database="Cafe-Datenbank")

# Cursor erstellen
cursor = conn.cursor()

# Log-Tabelle für das Cafe erstellen
cursor.execute('''
    CREATE TABLE IF NOT EXISTS log_cafe (
        id SERIAL PRIMARY KEY,
        produkt TEXT,
        produkt_id INTEGER,
        zeitstempel TIMESTAMP,
        aenderungsart TEXT
    )
''')


# Transaktion bestätigen
conn.commit()

# Verbindung schließen
cursor.close()
conn.close()
```
