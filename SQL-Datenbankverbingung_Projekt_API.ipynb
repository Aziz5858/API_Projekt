```python
import psycopg2

# Verbindung zur PostgreSQL-Datenbank herstellen
conn = psycopg2.connect(
    dbname="Bücher-Datenbank",
    user="postgres",
    password="Triplem94",
    host="localhost",
    port="5432"
)

# Cursor erstellen
cur = conn.cursor()

# Tabelle für Bücher erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS buch (
        id SERIAL PRIMARY KEY,
        titel TEXT,
        autor TEXT,
        erscheinungsjahr INTEGER,
        herausgeber TEXT,
        seitenanzahl INTEGER,
        preis NUMERIC,
        auf_lager BOOLEAN
    )
''')

# Tabelle für E-Books erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS ebook (
        id SERIAL PRIMARY KEY,
        titel TEXT,
        autor TEXT,
        erscheinungsjahr INTEGER,
        herausgeber TEXT,
        seitenanzahl INTEGER,
        preis NUMERIC
    )
''')

# Tabelle für Filme erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS film (
        id SERIAL PRIMARY KEY,
        titel TEXT,
        produzent TEXT,
        erscheinungsjahr INTEGER,
        preis NUMERIC,
        auf_lager BOOLEAN
    )
''')

# Transaktion bestätigen
conn.commit()

# Verbindung schließen
cur.close()
conn.close()
```

```python
import psycopg2

# Verbindung zur PostgreSQL-Datenbank herstellen
conn = psycopg2.connect(
    dbname="Cafe-Datenbank",
    user="postgres",
    password="Triplem94",
    host="localhost",
    port="5432"
)

# Cursor erstellen
cur = conn.cursor()

# Tabelle für Backstock erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS backstock (
        id SERIAL PRIMARY KEY,
        kuhmilch INTEGER,
        hafermilch INTEGER,
        kaffeebohnen INTEGER,
        limonade INTEGER,
        apfelschorle INTEGER,
        wasser INTEGER
    )
''')

# Tabelle für Latte erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS latte (
        id SERIAL PRIMARY KEY,
        preis NUMERIC,
        bestelldatum DATE
    )
''')

# Tabelle für Americano erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS americano (
        id SERIAL PRIMARY KEY,
        preis NUMERIC,
        bestelldatum DATE
    )
''')

# Tabelle für Limonade erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS limonade (
        id SERIAL PRIMARY KEY,
        preis NUMERIC,
        bestelldatum DATE
    )
''')

# Tabelle für Apfelschorle erstellen
cur.execute('''
    CREATE TABLE IF NOT EXISTS apfelschorle (
        id SERIAL PRIMARY KEY,
        preis NUMERIC,
        bestelldatum DATE
    )
''')

# Transaktion bestätigen
conn.commit()

# Verbindung schließen
cur.close()
conn.close()
```
